```python
import numpy as np #as np buna kısa isim veriyoruz.
import matplotlib.pyplot as matplot #verinin görselleştirmesi için bu kütüphane kullanılıyor

# numpay:verilerin oluşturulmasına yardımcı oluyor
# matplotlib:verinin görselleştirmesine yardımcı oluyor
```

```python
maasListesi=np.random.normal(4000,500,1000) #numpy kütüphanesinden rastgele getir normal dağılım olarak getir
#numpy.random.normal(4000,500,1000)->4 bin lira civarında 500 lira standart sapmalı 1000 tane veri oluştur
np.mean(maasListesi)#maasListesi nin ortalamasını al
```

```python
#hist:histogram
matplot.hist(maasListesi,50) #50 tane histogram yapsa yeterli
matplot.show() #bunları listeledi
```
