```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#grafikleri kolayca görmeye yarıyor
%matplotlib inline
```

```python
yasListesi=[10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi=[20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi=np.array(yasListesi)
numpyKiloListesi=np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
#çizim işlemi yaptırma->"g":green yeşil,"r":kırmızı xlabel:x ekseninin etiketi
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
#plt.show() :buraya yazsakta yazmasakta görsel gözükecek çünkü yukarıda tanımladık
#grafiğe başlık koymak
plt.title("Kilo'nun Yaşa Göre Değişim Grafiği")
```

```python
numpyDizisi=np.linspace(0,10,20)
```

```python
numpyDizisi
```

```python
numpyDizisi2=numpyDizisi**3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi,numpyDizisi2,"b-*")
#g--:kesikli olmasına yarıyor
#g*-:yıldızlarla göstermeye başlıyor
```

```python
#alt grafik:subplot ilk dizin kaç tane sıra olcağı ikinci dizin de kaç tane grafik
#son dizin hangisini çizecek 
plt.subplot(1,2,1)
plt.plot(numpyDizisi,numpyDizisi2,"r*-")
plt.subplot(1,2,2) #1 sıra olacak,2 kolon olacak, şu an ikinci grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi,"g--")
```

```python
benimFigure=plt.figure()
```

```python
benimFigure=plt.figure()
figureAxes=benimFigure.add_axes([0.4,0.4,0.6,0.6]) #0.9:genişlik,0.2:uzunluk tablonun küçülüp büyülmesine etkili son iki floatlı sayı
#argümanları liste şeklinde vermemiz lazım geniliğini ve uzunluğunu giriyorsun
figureAxes.plot(numpyDizisi,numpyDizisi2,"g")
#x ve y eksenlerinin isimlerini ver
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

# iç içe grafikler

```python
figure2=plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.2,0.4,0.3,0.3]) #ilk ik, kısım kenarlara uzaklığı belirliyor

eksen1.plot(numpyDizisi,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlığı")
```

```python
plt.subplots()
```

```python
(benimFigure,benimEksen)=plt.subplots(nrows=1,ncols=2)
benimEksen.plot(numpyDizisi,numpyDizisi2,"y") #hata alırız çünkü iki eksen oldu
```

# subplots

```python
(benimFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2)
type(benimEksenler)
```

```python
(benimFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2)
#olması için for kullanmalıyız

for eksen in benimEksenler:
    eksen.plot(numpyDizisi,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")

#arasındaki mesafeyi açalım
plt.tight_layout()

#benimFigure  #:iki defa yazdırır toplam 4 olur
```

```python
#yeniFigur = plt.figure(figsize=(3,3)) #figure içine boyutunu yazıp ayarlıyabiliyoruz
yeniFigur = plt.figure(dpi=100)

yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi,numpyDizisi**2,label="numpyDizisi**2") #hangi veri setini tanıdığını veriyor
yeniEksen.plot(numpyDizisi,numpyDizisi**3,label="numpyDizisi**3")
yeniEksen.legend(loc=2) #içine yeerini belirlemek için loc yazıyorsun
```

```python
yeniFigur.savefig("benimfigur.png",dpi=200) #kaydediyor klasöre
```
